## Clean Missing Values


## anime dataframe


```
clean_anime = anime_df.replace({'Unknown':np.nan})
clean_anime.dropna(inplace=True)
clean_anime.reset_index(inplace=True, drop=True)
```

```
clean_anime
```

## animelist dataframe


```
clean_animelist = animelist_df.rename(columns={'anime_id':'MAL_ID'})
```

```
clean_animelist
```

## anime_wiht_synopsis dataframe

```
clean_anime_with_synopsis = anime_with_synopsis_df.replace({'Unknown':np.nan})
clean_anime_with_synopsis.dropna(inplace=True)
clean_anime.reset_index(inplace=True, drop=True)
```

```
clean_anime_with_synopsis
```

## Entity Resolution


Need to split the genre coloumn for clean_anime to be able to filter on genre

```
clean_anime['Genres'] = clean_anime.apply(lambda x: [j.strip() for j in x['Genres'].split(",")],axis=1)
```

```
clean_anime = clean_anime.explode('Genres')
```

```
clean_anime
```

## Export Data



```
clean_anime.to_csv("anime_clean.csv", index=False)
clean_animelist.to_csv("animelist_clean.csv", index=False)
clean_anime_with_synopsis.to_csv("anime_with_synopsis_clean.csv", index=False)
```
